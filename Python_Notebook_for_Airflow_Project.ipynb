```python
genre_list = []
sub_genre_lst = []

for genre in top1960_df["spotify_genre"]:
    if type(genre) == str and genre != "[]":
        genre = genre.replace("[", "")
        genre = genre.replace("]", "")
        genre = genre.replace("'", "")
        #genre = genre.replace(" ", "")
        x = genre.split(",")
        genre_list.append(x)
        
for genre in genre_list:
    for sub_genre in genre:
        sub_genre_lst.append(sub_genre)
        
genre_pd_1960s = pd.DataFrame(sub_genre_lst)
genre_pd_1960s = genre_pd_1960s.rename(columns={0: "genre"})
genre_pd_1960s = genre_pd_1960s['genre'].str.lstrip(" ")

genre_pd_1960s.value_counts()  
```
